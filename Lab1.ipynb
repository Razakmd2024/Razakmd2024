```python
import pandas as pd
from pandas import read_html
import html5lib
import matplotlib.pyplot as plt
import seaborn as sns
Url="http://ubdmis.telangana.gov.in/BIRTHDEATH_UPENMAND.do?pstcode=35&style=bluetheme&distcode=137"
table=pd.io.html.read_html(Url)
ubd=table[2]
df=ubd.copy()
df.columns=df.columns.droplevel(0)
df.drop("Sl.No",axis=1,inplace=True)
a=df["Total No. of GPs in the District"]
b=df["Total No. of GPs Issuing Digitally Signed Certificates (Mee-Seva)"]
df["% of Digital Token mapping"]=round(b/a*100,2)
df1=df.sort_values("% of Digital Token mapping", ascending=False)
#rint(df1)
df1.drop(13,inplace=True)
df1.to_excel("Digital_Tokens.xlsx")
plt.figure(figsize=(16,8)) 
sns.barplot(x="Mandal Name", y="% of Digital Token mapping", data=df1,palette="Set2")
sns.scatterplot(x="Mandal Name", y="% of Digital Token mapping", data=df1, palette="Set2_r")
for i, k in zip(df1['Mandal Name'], df1['% of Digital Token mapping']):
    plt.annotate(k,(i,k), xytext=(5,-5), textcoords='offset points ', fontsize=12)

plt.xlabel("Mandal Name", fontsize=26)
plt.ylabel("% of Digital Token Mapping", fontsize=12)

plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=14)

plt.title("Digital Token Mapping Status in UBD portal of Peddapalli Dist ", fontsize=14)
plt.tight_layout()
plt.savefig("DigitalMapping.jpeg")
plt.show()
```
