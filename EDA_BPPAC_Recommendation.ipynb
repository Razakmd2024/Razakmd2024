# Project Description & Requirement 


## National PANCHAYAT AWARDS-2023

Ministry of Panchayati Raj,Govt. of India, invited all Gram Panchayats for nomination of National Panchayat Awards 2023 through online(panchayataward.gov.in) mode. The Nomination process at Gram Panchayat level over on 15-11-2022. 
Now, Block Level recommendation process is going. 
The Block Panchayat Committee need to verify all Gram Panchayat records in 9 themes , after field verification , Committee has to recommend  top 3 (best) Gram Panchayat's to next level for further process within the time schedule. 

## Requirement 
As a District Project Manager, we need to monitor status of recommendations in each theme of each Mandal of our District. 

Our District Contains 13 Mandals, 266 Gram Panchayat. To monitor the progress of each Mandal i need to open each Mandal url separently and note down status of recommendation. 

It's time consumming process.. 
I decide to write python script to easy my process by using Pandas, Numpy and Matplotlib etc...

For this Project Krish Naik Video's helped me a lot.. Thanks to him.

## Procedure 

1. Extractring the data from National Panchayat Award Dashbord 
2. Select the Table for DataFrame 
3. Remove the Unnecessary rows and Columns (Clean the Data)
4. Assign the New Headers to DataFrame 
5. Repeat the same process to remaining Mandals
6. Consolidate the all Mandals to one File
7. Prepare the GP Wise Recommended List and save in Excel
8. Prepare the Mandal Wise List and Save in Excel
9. Analysis the data...
